# **Problem Statement: Grid Search**

Load the‘voice.csv’ dataset intoa DataFrameand perform the following tasks:

1.Considering the ‘label’ column as the target variable, rename the column as ‘Gender_Identified’

2.Using the preprocessing()function,label the target column

3.Separatethe target variable and the feature vectors

4.Builda RandomForestClassifier model and find the best parameters usinga Grid search  

5.Print the best parameters and the best estimator

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/voice.csv')
df.head()
```

```python
df.shape
```


**1.Considering the ‘label’ column as the target variable, rename the column as ‘Gender_Identified’**


```python
df.rename(columns={'label': 'Gender_Identified'}, inplace=True)
df.head()
```


**2.Using the preprocessing()function,label the target column**


```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.dtypes
```

```python
en=LabelEncoder()
df['Gender_Identified']=en.fit_transform(df['Gender_Identified'])
df.head()
```


**3.Separatethe target variable and the feature vectors**


```python
x=df.drop(['Gender_Identified'],axis=1)
y=df.Gender_Identified
display(x)

display(y)
```


**4.Build a RandomForestClassifier model and find the best parameters usinga Grid search**  


```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
params = { "criterion" : ["gini", "entropy"], "n_estimators": [100, 150, 200,300]}
rf_gsv=GridSearchCV(estimator=RandomForestClassifier(),param_grid=params,cv=3,scoring='accuracy')
rf_gsv.fit(x,y)
```

```python
pd.DataFrame(rf_gsv.cv_results_).sort_values('rank_test_score')
```


**5.Print the best parameters and the best estimator**

```python
print('The best parameters are:')
rf_gsv.best_params_
```

```python
print('The best estimator is:')
rf_gsv.best_estimator_
```
